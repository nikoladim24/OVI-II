```python
import pandas as pd
```

```python
df = pd.read_csv('data/fruits_train.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['classification']
x = df.iloc[:,:-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 50, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion = 'entropy')
clf.fit(x_train,y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)*100
```

```python
df_high = df[df['calories (per 100g)'] >= 50]
df_low = df[df['calories (per 100g)'] < 50]
```

```python
y_high = df_high['classification']
x_high = df_high.iloc[:,:-1]
```

```python
y_low = df_low['classification']
x_low = df_low.iloc[:,:-1]
```

```python
y_predict_high = clf.predict(x_high)
y_predict_low = clf.predict(x_low)
```

```python
accuracy_score(y_predict_high,y_high)*100
```

```python
accuracy_score(y_predict_low,y_low)*100
```
